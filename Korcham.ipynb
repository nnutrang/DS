```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import googletrans
import time
from googletrans import Translator
# define a translate object
translator = Translator()
```

```python
def parse_detail_title():
    table =[]
    url='http://kocham.kr/content/company/zone_view.php?no=16'
    response = requests.get(url)
    soup = BeautifulSoup(response.content,"html.parser")
    
    board = soup("div",class_="board_view")
    board_tag = board[0].find_all('tr')
    #Parse tiêu đề
    title=[]
    for i in board_tag:
        title.append(translator.translate(i.text.split('\n')[1],dest = 'en').text)
    table.append(title)
    return table
```

```python
table = parse_detail_title()
```

```python
def parse_detail_board(table): 
    for i in range(2526,2681):
        url=f'http://kocham.kr/content/company/zone_view.php?no={i}'
        response = requests.get(url)
        soup = BeautifulSoup(response.content,"html.parser")
        
        board = soup("div",class_="board_view")
        board_tag = board[0].find_all('tr')
        #Parse nội dung
        row=[]
        for i in board_tag:
            try:
                row.append(translator.translate(i.text.split('\n')[2],dest = 'en').text)
            except:
                row.append(i.text.split('\n')[2])
        table.append(row)

    return table
```

```python
table = parse_detail_board(table)
```

```python
def parse_content(table):
    area_list = []
    name_list=[]
    address_list = []
    contact_list=[]
    sign_up_date_list=[]
    due_date_list=[]
    for i in range(1,233):
        url = f'https://www.korchamvietnam.com/service/directory/company/list?fbclid=IwAR136i7N3gs3UvCrDE8q0VNocRV8JecAMopNRQm-vW5jaYYc0bH8lyojMkM&cp={i}'
        response = requests.get(url)
        soup = BeautifulSoup(response.content,"html.parser")
        
        board = soup("div",class_="t-type02")
        board_tag=board[0].find_all('tr')
        for index in range(0,len(board_tag),2):
            temp_1 = board_tag[index].find_all('td')
            temp_2 = board_tag[index+1].find_all('td')
            
            area_list.append(temp_1[0].text)
            contact=list(filter(None,temp_1[2].text.split('\n')))
            for i in range(len(contact)):
                contact[i]=contact[i].strip()
            contact = ' '.join(contact)
            contact_list.append(contact)
            

            name = temp_1[1].text.split('\n')
            name = list(filter(None,name))
            for i in range(len(name)):
                name[i] = translator.translate(name[i],dest='en').text
            name = '-'.join(name)
            
            name_list.append(name)
            
            address = temp_2[0].find('div',class_='block-add').text.split('\n')
            address = list(filter(None,address))
            for j in range(len(address)):
                address[j]=address[j].strip()
            address = " ".join(address)
            address_list.append(address)
            try:
                korcham_member = temp_2[0].find('div',class_='korcham_member').text.split('\n')
                korcham_member = list(filter(None,korcham_member))
                sign_up_date_list.append(korcham_member[1].strip())
                try:
                    due_date_list.append(translator.translate(korcham_member[2].strip(),dest='en').text)
                except:
                    due_date_list.append(' ')
            except:
                sign_up_date_list.append(' ')
                due_date_list.append(' ')
            
    table['Area'] = area_list
    table['Name'] = name_list
    table['Address'] = address_list
    table['Contact'] = contact_list
    table['Sign_up_date'] = sign_up_date_list
    table['Due_date'] = due_date_list
    return table
```
